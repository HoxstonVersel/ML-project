```python
import requests
from bs4 import BeautifulSoup as bs
import csv
import pandas as pd
import numpy as np
```

```python
headers = {
    'accept':'*/*',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'
}

base_url = 'https://www.kn.kz/almaty/prodazha-odnokomnatnyh-dvuhkomnatnyh-trehkomnatnyh-chetyrehkomnatnyh-pyatikomnatnyh-kvartir/1/'

def kn_parse(base_url,headers):
    jobs = []
    session = requests.Session()
    request = session.get(base_url, headers = headers)
    if request.status_code == 200:
        soup = bs(request.content, 'html.parser')
        divs = soup.find_all('li', attrs = {'class':'results-item object-row'})
        print(f'Status {request.status_code}')
#         print(f'divs  {divs}')
        for div in divs:
            reklama = div.find('span', attrs ={'class': 'check-the-price-text'})
            if reklama is not None:
                continue
            # print(f'div {div}')
            title = div.find('a', attrs = {'class':'results-item-street'}).text
            href = div.find('a', attrs = {'class':'results-item-street'})['href']
            address = title.split(', ')[1:]
            
            area_tmp = div.find('li', attrs = {'class':'results-item-square'}).find('span', attrs = {'class':'field-value'}).text
            area = area_tmp.split('/')[0] if area_tmp.find('/')!=-1 else area_tmp
            
            year_tmp = div.find('li', attrs = {'class':'results-item-build'}).find('span', attrs = {'class':'field-value'}).text
            year = year_tmp.split('/')[0]
            
            price = div.find('p', attrs = {'class':'results-item-price'}).text.strip().split('/')[0]
            if price.find('Срочно!\r\n '):
                price = price.split('\n')[len(price.split('\n'))-1].strip()
            price = price.strip().split(' ₸', 1)
            content = div.find('p', attrs = {'class':'results-item-descr'}).text.strip()
            user_data = div.find('div', attrs = {'class':'results-item-authot-user-data'}).find_all('noindex')
#             print(price)
            
            
            jobs.append({
                'title': title,
                'href': href,
                'address': address,
                'year': year,
                'price': price,
                'content': content,
                'area': area
            })
    else:
        print('ERROR')
#     print(jobs)
    return jobs
        
def files_writer(flats):
    #with open('HeadHunter.csv', 'a', encoding='utf-8') as file:
    with open(r"KNKZ#4.csv", "a", encoding = 'utf-8') as file:
        a_pen = csv.writer(file)
        a_pen.writerow(('title', 'price', 'href','address','year','content','area'))
        for flat in flats:
            a_pen.writerow((flat['title'],int(flat['price'][0].replace(" ", '')),flat['href'],flat['address'],flat['year'],flat['content'],flat['area']))

        


 

flats = kn_parse(base_url, headers)
files_writer(flats)
```
